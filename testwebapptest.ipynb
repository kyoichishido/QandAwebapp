```python
from transformers import DistilBertTokenizer, DistilBertForQuestionAnswering
import torch
```

```python
tokenizer = DistilBertTokenizer.from_pretrained('distilbert-base-uncased',return_token_type_ids = True)
model = DistilBertForQuestionAnswering.from_pretrained('distilbert-base-uncased-distilled-squad')
```

```python
#para = widgets.Text(placeholder='write a paragraph')
para = 'I have a yellow colour car'
```

```python
#que = widgets.Text(placeholder='ask you questions')
que = widgets.Text(
    placeholder='Type something',
    description='Question:',
    disabled=False
)
```

```python
button_send = widgets.Button(
                description='Send to system',
                tooltip='Send',
                style={'description_width': 'initial'}
            )

output = widgets.Output()

def on_button_clicked(event):
      with output:
        clear_output()
        print({que.value})
        print(type({que.value}))
        print(type(que))
        encoding = tokenizer.encode_plus(que.value, para)
        input_ids, attention_mask = encoding["input_ids"], encoding["attention_mask"]
        start_scores, end_scores = model(torch.tensor([input_ids]), attention_mask=torch.tensor([attention_mask]))
        ans_tokens = input_ids[torch.argmax(start_scores) : torch.argmax(end_scores)+1]
        answer_tokens = tokenizer.convert_ids_to_tokens(ans_tokens , skip_special_tokens=True)
        print ("\nQuestion ",{que.value})
        print ("\nAnswer Tokens: ")
        print (answer_tokens)

        answer_tokens_to_string = tokenizer.convert_tokens_to_string(answer_tokens)

        print ("\nAnswer : ",answer_tokens_to_string)
                
button_send.on_click(on_button_clicked)
vbox_result = widgets.VBox([button_send, output])
```

```python
text_0 = widgets.HTML(value="<h1>Question answering System!</h1>")
text_1 = widgets.HTML(value="<h2>Ask your question</h2>")
vbox_text = widgets.VBox([text_0, text_1, que, vbox_result])
```

```python
page = widgets.HBox([ vbox_text])
display(page)
```
